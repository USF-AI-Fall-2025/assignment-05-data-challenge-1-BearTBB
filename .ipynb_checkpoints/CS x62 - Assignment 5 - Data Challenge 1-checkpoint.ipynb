# Null checking data

```python
df.info()
```

Some missing data for column DISTRICT_CODE. All other columns seem to be non-null for all rows.

Column numbers 0-6 seem to be of object Dtype, except for DISTRICT_CODE at column number 2, which is of Dtype float64.
This is peculiar, because it does not seem like DISTRICT_CODE would need to be a float64.
From my understanding of the data, DISTRICT_CODE is supposed to be a discrete, integer representation of the DISTRICT_NAME column.
I will verify this information using value_counts().

```python
df.DISTRICT_CODE.value_counts()
```

Indeed, there are multiple instances of each number. Let's see if each number is associated with a specific DISTRIC_NAME.

```python
df.value_counts(["DISTRICT_NAME", "DISTRICT_CODE"])
```

The length of the created table seems to be equal for each value_counts(),
which suggests that each existing DISTRICT_CODE maps exactly to one DISTRICT_NAME.

This also suggests that the null values for DISTRICT_CODE is because some DISTRICT_NAMEs are not yet assigned an official code.

I'll be changing the Dtype of DISTRICT_CODE to object to make it easier to work with,
and also make it behave more closely to how the data is actually treated.

```python
df.DISTRICT_CODE = df.DISTRICT_CODE.astype(str).str.split('.', expand = True)[0]
df
```

```python
sns.displot(df.WAGE_YEAR1[df.WAGE_YEAR1 > 0], bins=10, alpha=1)
sns.displot(df.WAGE_YEAR2[df.WAGE_YEAR2 > 0], bins=10, alpha=1)
sns.displot(df.WAGE_YEAR3[df.WAGE_YEAR3 > 0], bins=10, alpha=1)
sns.displot(df.WAGE_YEAR4[df.WAGE_YEAR4 > 0], bins=10, alpha=1)
```

# Making the Model

I have identified quite a few number of features that I think are not quite useful in the making of this model. I'll drop them in this stage, and store the result in a new dataframe

```python
df_train = df.drop(['DISTRICT_CODE', 'ACADEMIC_YEAR', 'DEMO_CATEGORY'], axis=1)
```

Next, I will encode all of the values that are not in number form in order to be able to be fed into the model. I will use label encoding for all of the data points, although I recognize that one-hot encoding might have been better here for DISTRICT_TYPE and AWARD_CATEGORY.

```python
from sklearn.preprocessing import LabelEncoder
labels = LabelEncoder()
df_train['DISTRICT_TYPE'] = labels.fit_transform(df_train['DISTRICT_TYPE'])
df_train['DISTRICT_NAME'] = labels.fit_transform(df_train['DISTRICT_NAME'])
df_train['STUDENT_POPULATION'] = labels.fit_transform(df_train['STUDENT_POPULATION'])
df_train['AWARD_CATEGORY'] = labels.fit_transform(df_train['AWARD_CATEGORY'])
df_train
```

Next, I would like to verify that my chosen algorithm could get a RSME of ~2500-3000. I will do this by splitting the training data into test and train data sets, so that I have something to test against.

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_train.drop(['WAGE_YEAR4'], axis = 1), df_train.WAGE_YEAR4, test_size = 0.25)
```

My chosen algorithm for this problem is linear regression, because of ease of use and the fact that we are trying to predict a continuous value.

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(train_x, train_y)
```

Oh wow, I didn't actually expect the RMSE to be that low. This is great news. I hope to have good results when comparing with the actual test dataset.

```python
from sklearn.metrics import root_mean_squared_error

rms = root_mean_squared_error(test_y, regression.predict(test_x))
rms
```
